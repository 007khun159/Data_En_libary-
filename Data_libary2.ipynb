```python
#Concatenate 
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3'],
                    'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']},index =[0,1,2,3])
df1
```

```python
df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
                    'B':['B4','B5','B6','B7'],
                    'C':['C4','C5','C6','C7'],
                    'D':['D4','D5','D6','D7']},index =[0,1,2,3])
df2
```

```python
df3 = pd.DataFrame({'A':['A7','A8','A9','A10'],
                    'B':['B7','B8','B9','B10'],
                    'C':['C7','C8','C9','C10'],
                   'D':['D7','D8','D9','D10']},index =[0,1,2,3])
df3
```

```python
left = pd.DataFrame(
    {
        "key1": ["K0", "K0", "K1", "K2"],
        "key2": ["K0", "K1", "K0", "K1"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
    }
)
```

```python
right = pd.DataFrame(
    {
        "key1": ["K0", "K1", "K1", "K2"],
        "key2": ["K0", "K0", "K0", "K0"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    }
)
```

```python
result = pd.merge(left, right, on=["key1", "key2"])
result
```

```python
result = pd.merge(left, right, how="left", on=["key1", "key2"])
result
```

```python
result = pd.merge(left, right, how="outer", on=["key1", "key2"])
result
```
